**How would you convert tweets into numerical representations (TF-IDF,
CountVectorizer, Word2Vec)?**

```python
x = data['text']
y = data['sentiment']
x
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(x)
print(vectorizer.get_feature_names_out())
print(X.toarray())
```

# **You are training a Random Forest model for sentiment classification.**

```python
from sklearn.metrics import accuracy_score,classification_report
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=42)
```

```python
classifier = RandomForestClassifier()
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
print(classification_report(y_test,y_pred,target_names = ['negative','positive']))
```

# **Which hyperparameters are crucial, and how would you optimize them using GridSearchCV or RandomizedSearchCV**

**We will use RandomizedSearchCV because it is used for large dataset**

```python
from sklearn.model_selection import RandomizedSearchCV

param_dist = {
    'n_estimators': np.arange(50, 300, 50),  # Number of trees in the forest
    'max_depth': [None, 10, 20, 30, 40, 50],  # Maximum depth of each tree
    'min_samples_split': np.arange(2, 10, 2),  # Minimum samples required to split a node
    'min_samples_leaf': np.arange(1, 5, 1),  # Minimum samples required at a leaf node
    'criterion': ['gini', 'entropy']  # Function to measure the quality of a split
}
```

```python
random_search = RandomizedSearchCV(
    estimator=classifier,  # The model to optimize
    param_distributions=param_dist,  # The hyperparameter grid
    n_iter=20,  # Number of random combinations to try
    cv=5,  # 5-fold cross-validation
    scoring='accuracy',  # Metric to evaluate model performance
    n_jobs=-1,  # Use all available CPU cores
    random_state=42  # Ensure reproducibility
)
```

```python
random_search.fit(X_train, y_train)

print("Best Hyperparameters:", random_search.best_params_)
```

# **How does each architecture handle CNN & LSTM sequential text data?**


In CNN,Convolution is used in feature extraction & it is used in image and text precessing and the pooling layer is used to reduce the dimentionality and keep the most relevant feature.If we work with short text and need speed we can work with CNN.the flatenning and dense layer of CNN helps us in classification.


```python
!pip install tensorflow
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, LSTM, Dense, Dropout
```

```python
embedding_layer = Embedding(input_dim=5000, output_dim=128, input_length=50)

# CNN Model
cnn_model = Sequential([
    embedding_layer,
    Conv1D(filters=128, kernel_size=5, activation='relu'),
    GlobalMaxPooling1D(),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

cnn_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

print("CNN Model Summary:")
cnn_model.summary()
```

```python
embedding_layer = Embedding(input_dim=5000, output_dim=128, input_length=50)

lstm_model = Sequential([
    embedding_layer,
    LSTM(128, return_sequences=False),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

lstm_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

lstm_model.summary()
```

#**How does BERT handle contextual word meanings compared to Word2Vec or TF-IDF?**

BERT, Word2Vec, and TF-IDF handle contextual word meanings differently. BERT provides dynamic embeddings, adapting word meanings based on context with a bidirectional approach, making it ideal for complex sentences and polysemy.Word2Vec offers static embeddings, giving each word a fixed vector regardless of context, which is suitable for semantic similarity but lacks contextual sensitivity.TF-IDF uses a frequency-based approach, focusing on word importance in documents without considering context or word order, making it useful for simple keyword extraction but not for deep understanding.
